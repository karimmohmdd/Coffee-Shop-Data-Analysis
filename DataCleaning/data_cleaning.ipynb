## Problem 3: Money Outliers

```python
df[df['money'] > 100]
```

```python
from thefuzz import process
```

```python
Q1 = df['money'].quantile(0.25)
Q3 = df['money'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + (1.5 * IQR)
lower_bound = Q1 + (1.5 * IQR)
upper_bound
```

```python
df[df['money'] > upper_bound].describe()
```

```python
df[df['money'] > 100].groupby('coffee_name').count().money
```

```python
correct_names = df['coffee_name'].value_counts().head(8).index.tolist()
correct_names
```

```python
df['money'] = df['money'].astype(str)
df['money'] = df['money'].str.replace('-' , '')
df['money'] = df['money'].astype(float)
```

```python
from thefuzz import process
```

```python
correct_names = df['coffee_name'].value_counts().head(8).index.tolist()
correct_names
```

```python


def correct_name(x):
    if pd.isna(x):
        return x
    match, score = process.extractOne(x, correct_names)
    return match if score > 70 else x

df['coffee_name'] = df['coffee_name'].apply(correct_name)
```

```python
df['coffee_name'].value_counts()
```

```python
means = df[df['money'] < upper_bound].groupby('coffee_name').money.mean()
means
```

```python
df[df['money'] > 50]
```

```python
for i in df.index:
    if df.loc[i , 'money'] > 100:
        df.loc[i , 'money'] = np.nan

df[df['money'] > 40].groupby('coffee_name').money.count()
```

```python
df.sort_values(by='money' , ascending = False)
```

```python
df['money'] = df['money'].fillna(df['coffee_name'].map(means))
df.sort_values(by='money' , ascending = False)
```

```python
df = df.dropna(subset=['money', 'coffee_name'], how='all')
df
```

## Problem 8: Time_of_Day

```python
df['Time_of_Day'].value_counts()
```

```python
correct_time_names = df['Time_of_Day'].value_counts().head(3).index.tolist()
correct_time_names
```

```python
def correct_name1(x):
    if pd.isna(x):
        return x
    match, score = process.extractOne(x, correct_time_names)
    return match if score > 70 else x
df['Time_of_Day'] = df['Time_of_Day'].apply(correct_name1)
```

```python
df['Time_of_Day'].value_counts()
```

```python
df[df['Time_of_Day'] == 'Night'].sort_values(by = 'hour_of_day')
```

```python
time_median = df.groupby('Time_of_Day').hour_of_day.median()
time_median
```

```python
df['hour_of_day'] = df['hour_of_day'].fillna(df['Time_of_Day'].map(time_median))
df
```

```python
def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 23:
        return 'Night'

df['Period'] = df['Hour'].apply(get_time_of_day)
```

```python
df['money'] = df['money'].round(2)
```

```python
df.rename(columns={'hour_of_day': 'Hour' , 'Date_fixed': 'Date' , 'Month_name': 'Month' , 'Time_of_Day': 'Period'}, inplace=True)
```
